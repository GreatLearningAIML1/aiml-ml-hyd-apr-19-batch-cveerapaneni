```python
import numpy as np
import pandas as pd

# Read the CSV file. Use the 2nd row as the header (as the first row has empty data)
# Consider "-" as NaN value. 
team_data = pd.read_csv('../data/Laliga.csv', header=1, na_values='-')
# Now replace NaN with 0
team_data.fillna(0, axis=1, inplace=True)
# past_data.dtypes
team_data
```

```python
# This method takes in the teams data, extracts out the 'Debut' data (list).
# Filters out all rows that do not fall between the provided "debut_from" and "debut_to".
def find_teams_who_debuted_between(team_data, debut_from, debut_to):
    # Extract the Debut column.
    team_debut_data = team_data['Debut'];
    
    # Empty array to hold the decisions (True / False) indicating if the row should be
    # included in the final subset or not.
    filtered_rows = []
    
    for team_debut in team_debut_data:
        # Data type of 'Debut' column is Object. Convert it into a string and then split by "-"
        # as in some cases, the debut value is shown as a range.
        debut_year_duration = str(team_debut).split('-')
        
        if len(debut_year_duration) > 2:
            # This case shouldn't arise as the duration is expressed in one of the below formats:
            # 1930, 1930-32, 2017-18, etc.
            print("Don't know how to handle...")
        elif len(debut_year_duration) == 2:
            # This could be something on the lines: 1930-32, 2017-18, etc.
            debut_duration_from = int(debut_year_duration[0])
            debut_duration_to = int(debut_year_duration[0][0:2] + debut_year_duration[1])
            filtered_rows.append(debut_duration_from >= debut_from and debut_duration_to <= debut_to)
        else:
            debut_year = int(debut_year_duration[0])
            filtered_rows.append(debut_year >= debut_from and debut_year <= debut_to)
    return team_data.loc[filtered_rows]

# Teams who have started playing between 1930 - 1980
print("List of teams who have debuted between 1930 and 1980 are shown below")
find_teams_who_debuted_between(team_data, 1930, 1980)
```

```python
# Find top 5 teams in terms of points
# Group the data by Team (to handle cases where there are multiple records for the same team)
grouped_by_team = team_data.groupby(by='Team')

print("--- Top 5 teams w.r.t points accumulated ---")
# Apply the aggregation operation on the grouped data
grouped_by_team['Points'] \
    .agg(np.sum) \
    .sort_values(ascending=False) \
    .head(5)
```

```python
# This method calculates the Goal difference count and returns the dataframe
# with an additional column that holds the goal difference count.
def Goal_diff_count(teams):
    # Calculate the goal difference and add to 'GoalDifferenceCount' column
    teams['GoalDifferenceCount'] = teams['GoalsFor'] - teams['GoalsAgainst']
    return teams

# Get the goal difference count and print it (last column in the data frame).
Goal_diff_count(team_data)

# Now sort the data in a descending mode by 'GoalDifferenceCount' column
# We do not want to do inplace sorting as we do not want to disturb the original dataset
team_data_sorted_by_gdf = team_data.sort_values(by = 'GoalDifferenceCount', ascending=False)

# First record in the data frame has the maximum goal difference count, while
# the last record in the data frame has the minimum goal difference count.
top_team = team_data_sorted_by_gdf.head(1)
top_team_name = top_team['Team'][0]
top_team_goal_diff = top_team['GoalDifferenceCount'][0]

# Calculate the index of the last row (after sorting)
last_row_index = team_data_sorted_by_gdf.index[len(team_data_sorted_by_gdf) - 1]
bottom_team = team_data_sorted_by_gdf.tail(1)
bottom_team_name = bottom_team['Team'][last_row_index]
bottom_team_goal_diff = bottom_team['GoalDifferenceCount'][last_row_index]

print("Team '{}' has maximum goal difference count of {}".format(top_team_name, top_team_goal_diff))
print("Team '{}' has minimum goal difference count of {}".format(bottom_team_name, bottom_team_goal_diff))

# Finally delete the 'GoalDifferenceCount' column
del(team_data['GoalDifferenceCount'])
```

```python
# Add a new column for Winning Percentage.

team_data['WinningPercent'] = (team_data['GamesWon'] / team_data['GamesPlayed']) * 100
team_data.fillna(0, axis=1, inplace=True)

print("--- Top 5 teams that have the highest winning percentage ---")
team_data.sort_values(by = 'WinningPercent', ascending=False).head(5)
```

```python
# Group teams based on their â€œBest positionâ€ and print the sum of their points for all positions

# We shall group by multiple attributes. Here, we are first grouping by 'Best Position' followed by 'Team'

# NOTE: Please see the below cell for another variation of first grouping by 'Team'
# followed by 'BestPosition'

print("--- Teams grouped by 'Best Position' and their 'Points' ---")
print("--- Dataset is grouped first on 'BestPosition' followed by 'Team' ---")

grouped_by_team_bp = team_data.groupby(by=['BestPosition', 'Team'])

# Apply the aggregation operation on the grouped data
grouped_by_team_bp[['Points']] \
    .sum() \
    .sort_values(by=['BestPosition', 'Points', 'Team'], ascending=[True, False, True])
```

```python
print("--- Teams grouped by 'Best Position' and their 'Points' ---")
print("--- Dataset is grouped first on 'Team' followed by 'BestPosition' ---")

# Another variation from the above is to first group by 'Team' followed by 'Best Position'
grouped_by_team_bp = team_data.groupby(by=['Team', 'BestPosition'])

# Apply the aggregation operation on the grouped data
grouped_by_team_bp[['Points']] \
    .sum() \
    .sort_values(by=['BestPosition', 'Points', 'Team'], ascending=[True, False, True])
```
